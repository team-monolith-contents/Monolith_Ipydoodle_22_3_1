# 1.2 물체의 여러가지 속성

각각의 물체에는 조절할 수 있는 여러가지 속성이 존재합니다.

예를 들어 `box` 의 경우 위치, 크기, 색 등을 조절할 수 있고, `circle`은 위치, 반지름, 색 등을 조절할 수 있습니다.

물체의 속성은 물체를 만들 때 다음과 같이 조절할 수 있습니다.

```python
Box(x = 100, y = 200, width = 50, height = 200, color = [255,0,0])
```

여기서 x, y 값은 각각 물체 중심의 x,y 좌표값을 의미하고 width, height는 각각 가로와 세로의 길이를 의미합니다.

마지막으로 color의 세 숫자는 각각 [빨강, 초록, 파랑] 의 정도를 나타냅니다.

````python
World()
Box(x = 100, y = 200, width = 30, height = 50, color = [255,0,0])
````

마찬가지의 방법으로 다양한 `circle`을 만들어 보도록 하겠습니다.

`circle`은 x, y, radius, color의 4가지 속성을 가지고 있습니다.

x, y는 각각 중심의 x ,y 좌표이고 radius 는 반지름 color는 색을 나타냅니다.

````python
World()
Circle(x = 0, y = 0, radius = 250, color = [255,0,0])
Circle(x = 0, y = 0, radius = 200, color = [0,255,0])
Circle(x = 0, y = 0, radius = 150, color = [0,0,255])
Circle(x = 0, y = 0, radius = 100, color = [0,0,0])
````

## 출력함수
---

print()를 사용해 한번에 많은 문장을 프린트 할 수도 있습니다.

print()안에 ,(반점)으로 구분하여 여러 문장들을 넣어주면 각각의 문장들이 space(공백)으로 구분되어 출력됩니다.

이 때, print()의 마지막에 sep="/"을 넣어주면 각각의 문장들을 space(공백)이 아닌 /로 구분하여 줍니다.

물론 sep="?"의 물음표 자리에는 어떠한 기호나 숫자들도 들어갈 수 있습니다.

```python
print("파이썬", "데이터분석입문", "PlaywithData", "데이터 수집")
```

```python
print("귤", "사과", "배", sep="/")
```
